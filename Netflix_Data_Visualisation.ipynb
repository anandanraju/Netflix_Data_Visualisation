```python
#Matplotlib - It is a low level graph plotting library that serves as a visualisation utility
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([0,6])
y=np.array([10,250])
plt.plot(x,y,color='red')
plt.show()
```

```python
#bar graph
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y,color='hotpink')
```

```python
#pie chart
slices = [50,25,45,10]
#Take department names
depts = ['Sales','Production','HR','Finance']
cols=['magenta','cyan','brown','gold']
#create pie chart
plt.pie(slices,labels=depts,colors=cols)
#set title and legend
plt.title('RIT')
plt.legend()
```

```python
#pie chart
slices = [50,25,45,10]
#Take department names
depts = ['Sales','Production','HR','Finance']
cols=['magenta','cyan','brown','gold']
#create pie chart
plt.pie(slices,labels=depts,colors=cols,explode=(0,0.2,0,0))
#set title and legend
plt.title('RIT')
plt.legend()
```

```python
x=np.array([0,6])
y=np.array([10,250])
plt.plot(x,y,color='red', marker='h',ms=10, mfc='c')
#Name of the marker is : 'h','s','o'
plt.show()
```

```python
x=np.array([1,3,4,5,6,7,8,9])
y=np.array([4,7,2,4,3,5,3,2])
plt.plot(x,y,color='violet',linewidth=7, marker='*',ms=10, mfc='m')
#Name of the marker is : 'h','s','o'
#Name of the mfc : 'magenta','cyan','brown','gold'
plt.show()
```
